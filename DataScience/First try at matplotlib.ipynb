```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[1,4,9], '-.s', color = (1,0.5,0.5, 0.5), markerfacecolor = 'y') #RGB + prozrachnost
x = np.array([1,1,5,5,1])
y = np.array([1,5,5,1,1])

y2 = np.arange(0,5,1)
x2 = np.array([a*a for a in y2])

plt.plot(x,y,'g--o', x2, y2, 'r:*')
#plt.plot(x3,y3)

plt.grid()

plt.show()
```

```python
plt.plot(x,y, 'None')

plt.show()
```

```python
x = np.arange(-2*np.pi,2*np.pi,0.1)
y = np.cos(x)
plt.plot(x,y)
plt.fill_between(x,y,0.05, where=(y<0), color = 'r', alpha = 0.3)
plt.fill_between(x,y,0.05, where=(y>0), color = 'g', alpha = 0.3)

plt.grid(color = 'y')
plt.show()
```

```python
from matplotlib.gridspec import GridSpec

ws = [1,2,5]
hs = [2, 0.5]

fig = plt.figure(figsize=(7,4))
gs = GridSpec(ncols=3,nrows=2,figure=fig, width_ratios=ws, height_ratios=hs)

ax1 = plt.subplot(gs[0,0])
ax1.plot(np.random.random(10))

ax2 = plt.subplot(gs[1,0:2])
ax2.plot(np.random.random(10))

ax3 = plt.subplot(gs[:,2])
ax3.plot(np.random.random(10))

plt.show()
```

```python
from matplotlib.ticker import NullLocator, LinearLocator,MultipleLocator
fig = plt.figure(figsize = (7,4))
ax = fig.add_subplot()
ax.plot(np.arange(0,15,1))

#ax.set(xlim=(-1,10), ylim=(-2,5))
ax.set_xlim(xmin=-1)
ax.set_ylim(-2,15)

#from matplotlib.ticker import NullLocator
ax.grid()
ax.xaxis.set_major_locator(LinearLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(1.5))

plt.show()
```

```python
from matplotlib.ticker import  IndexLocator,FixedLocator, LogLocator, MaxNLocator

fig = plt.figure(figsize = (7,4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.grid()
#ax.xaxis.set_major_locator(IndexLocator(base =0.5, offset = 0.57))
ax.xaxis.set_major_locator(LogLocator(base =2))
#ax.yaxis.set_major_locator(FixedLocator([-1,-0.5,0,1]))
ax.yaxis.set_major_locator(MaxNLocator(7))
```

```python
fig = plt.figure(figsize = (7,4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.minorticks_on()
ax.grid(which='major',lw = 2)
ax.grid(which='minor')
plt.show()
ax.xaxis.set_minor_locator(NullLocator())
```
