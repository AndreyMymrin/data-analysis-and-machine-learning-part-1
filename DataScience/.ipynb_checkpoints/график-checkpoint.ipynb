```python
import numpy as np
import pandas as pd
from IPython.display import display, Latex

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(palette='Set2', font_scale=1.3,style='white')
```

```python
y = np.arange(350,376,0.1)
#y = np.array([10.48,24.48])
#dataSet = np.array([x, y])
```

```python
Ha = 34900
Hb = 31800
R = 8.31
Ta = 374.4
Tb = 356.4
x = (1-np.exp(Hb/R*(1/Tb-1/y)))/(np.exp(Ha/R*(1/Ta-1/y))-np.exp(Hb/R*(1/Tb-1/y)))

gasx = x*np.exp(Ha/R*(1/Ta-1/y))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(7, 7))
plt.xlim([0, 1])
plt.ylim([0, 1])
ax.grid(color = 'grey')

ax.minorticks_on()
ax.grid(which='minor', color = 'grey',linestyle = ':')

plt.ylabel(r"$x_{gas}^a$")
plt.xlabel(r'$x_{liqud}^a$')
#plt.xlabel(r"sin($%s$)" %(r'\varphi'))
#xt = np.linspace(0, 2.2, 100)
#ax.errorbar(x, y, xerr = 2,fmt=' ',elinewidth=1,capsize=2,color = 'black') 
#plt.plot(xt,xt*180*3600/np.pi/np.sqrt(20000**2-xt**2*578**2),color = (0.3,0.3,0.3),linewidth = 0.8)
#print(1*180*3600/np.pi/np.sqrt(20000**2-1**2*578**2))
#plt.vlines(0.71, 82,105, color = 'r', linestyle = '-.')
plt.plot(x,gasx)
plt.show()
```
