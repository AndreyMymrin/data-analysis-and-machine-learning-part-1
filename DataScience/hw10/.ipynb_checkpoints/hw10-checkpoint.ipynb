Перед обучением линейной регрессии вы могли совершать некоторые преобразования над признаками. Какие преобразования стоит совершить перед обучением решающего дерева?

В частности, ответьте на вопрос, имеет ли смысл примененять следующие преобразования:
* монотонные функции от одного признака;

    Можно, например мы хотим избежать вычислительной ошибки. например если у нас  значения какого-то признака порядка $10^{-8}$, то можно домножить на 10^8, чтобы не было ошибки округления.


* немонотонные функции от одного признака;

    Можно в особых случаях, например в каком-то признаке мы храним какое-то отклонение в разные стороны и нам не важно, в какую сторону отклонение. Здесь мы можем использовать модуль.


* линейные функции от нескольких признаков;

    Можно, например когда хотим понизить размерность. Наприер мы знаем, что важна совокупная масса всех коробок, а не каждой коробки. Тогда $m_1,m_2$... можно заменить на M = $\Sigma m_i$
    

* нелинейные функции от нескольких признаков;

    Можно, например когда хотим понизить размерность. Например мы знаем, что важен объем коробки, а не 3 пространственные координаты. Тогда можно заменить a,b,c на V=abc


* стандартизация признаков;

    Не имеет смысла, попросту меняем пороговое значение. Тем более может потеряться интерпретируемость дерева.


* преобразования категориальных признаков.

    Мы это всегда. Например признак Male/Female мы преобразуем к 0/1, чтобы компьютер мог с этим работать, сравнивать и тд.

Например, вместо/дополнительно с признаком $x_j$ рассматривать признак $f(x_j)$, для указанных выше типов функций.


## Задача 2

### 1. Загрузка и предобработка данных.




Скачайте [данные](https://www.kaggle.com/datasets/sayansh001/crystal-structure-classification) по классификации кристаллических решеток. Целевой переменной в этой задаче является `Lowest distortion`. С сырыми данными и более подробным описанием задачи можно ознакомиться также [здесь](https://www.kaggle.com/datasets/sayansh001/crystal-structure-classification).


```python
data = pd.read_csv('Crystal_structure.csv')
data
```

Сделаем предобработку данных.

```python
# уберем признаки, которые нам не нужны (подробнее можете почиать во вкладке 'Code' на странице датасета)
data = data.drop(["In literature", "Compound", "τ"], axis=1)
# убираем объекты, для которых неизвестен таргет
data = data.drop(data[data["Lowest distortion"] == "-"].index).reset_index(drop=True)
# закодируем целевую переменную числами
data['Lowest distortion'] = LabelEncoder().fit_transform(data['Lowest distortion'])
# закодируем категориальные признаки числами
data['A'] = LabelEncoder().fit_transform(data['A'])
data['B'] = LabelEncoder().fit_transform(data['B'])
# обработаем пропуски
for i in data.columns:
    column = data[i]
    for k in range(len(column)):
        if column[k] == '-':
            column[k] = 0
    data[i] = column
data
```

### 2. Обучение модели


Создайте массив признаков и массив таргета.

```python
X = data[data.columns[:-1]]
y = data['Lowest distortion']
```

Обучите модель решающего дерева.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42, max_depth= 14, min_samples_leaf= 2)
clf.fit(X_train, y_train)
```

Сделайте предсказание для тестовой выборки.

```python
predictions = clf.predict(X_test)
print('test accuracy:', accuracy_score(predictions, y_test).round(3))
```

Создайте массив, показывающий успех или не успех модели (1 если модель угадала класс, 0 если не угадала).

```python
sample = predictions == y_test
```

Вы получили выборку из распределения $Bern(\theta)$. Постройте доверительный интервал для $\theta$.

$\widehat{\theta} = \overline{sample}$

$\widehat{\sigma} = \theta(1-\theta)$

$(\widehat{\theta} - z_{\frac{1+\alpha}{2}}\frac{\widehat{\sigma}}{\sqrt{n}},\widehat{\theta} + z_{\frac{1+\alpha}{2}}\frac{\widehat{\sigma}}{\sqrt{n}})$

```python
alpha = 0.95
z = sps.norm.ppf((1+alpha)/2)
n = sample.size
theta = sample.mean()
sigma = theta*(1-theta)
interval = np.array([theta-z*sigma/np.sqrt(n), theta+z*sigma/np.sqrt(n)])
interval.round(3)
```

Повторите все шаги для случайного леса. Сравните полученные интервалы.

```python
n_estimators = 300
model = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print('test accuracy:', accuracy_score(predictions, y_test).round(3))

sample = predictions == y_test
theta = sample.mean()
sigma = theta*(1-theta)
interval = np.array([theta-z*sigma/np.sqrt(n), theta+z*sigma/np.sqrt(n)])
interval.round(3)
```

**Выводы:** точность на тестовой выборке выше у случайного леса.

Дерево легко переобучается, поэтому пришлось подбирать гиперпараметры max_depth и min_samples_leaf, добиваясь наибольшей точности на тестовой выборке.

Для леса же переобучение желательно, поэтому не пришлось ничего подбирать, что делает его не только более точным инструментом, но и более удобным

```python
data = data0
i = 0
for column in data0.columns[:-2]:
    U1 = np.sort(data[column])[np.int(0.25*data.shape[0])]
    U3 = np.sort(data[column])[np.int(0.75*data.shape[0])]
    #print(U1,U3)
    data = data.drop(data[data[column] > U3+4.5*(U3-U1)].index).reset_index(drop=True)
    data = data.drop(data[data[column] < U1-4.5*(U3-U1)].index).reset_index(drop=True)
    i +=1
    #print(data.shape)
data.shape
```

```python
X, y = data.loc[:, "0":"299"], data['Eat']
X_not_test, X_test, y_not_test, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25)
```

Обучите решающее дерево глубины 3 и визуализируйте его. 

*Рекомендуется сделать скриншот визуализации и прислать его в бот вместе с ноутбуком.*

```python
clf = DecisionTreeRegressor(random_state=42, max_depth= 3)
clf.fit(X_train, y_train)
plt.figure(figsize=(15, 10))
tree.plot_tree(clf)
plt.show()
```

Подберите оптимальное значение максимальной глубины дерева. Для этого обучите решающие деревья с выставленным значением соответствующего гиперпараметра и посчитайте ошибку предсказаний каждого из них на валидационной выборке. Постройте соответствующий график.

```python
mape = []
depths = np.arange(1,20,1)
for depth in depths:
    clf = DecisionTreeRegressor(max_depth = depth, min_samples_leaf= 1,random_state=11)
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_valid)
    mape.append(mean_absolute_percentage_error(y_valid, predictions))
```

```python
plt.figure(figsize=(12, 3))
plt.plot(depths, mape)
plt.title('MAPE')
plt.xlabel('depth')
plt.ylabel('error')
```

Аналогичным образом подберите минимальное число объектов в листе, зафиксировав при этом подобранное значение максимальной глубины.

```python
mape = []
leafs = np.arange(1,20,1)
for leaf in leafs:
    clf = DecisionTreeRegressor(max_depth = 6, min_samples_leaf = leaf,random_state=2)
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_valid)
    mape.append(mean_absolute_percentage_error(y_valid, predictions))
```

```python
plt.figure(figsize=(12, 3))
plt.plot(depths, mape)
plt.title('MAPE')
plt.xlabel('min_samples_leaf')
plt.ylabel('error')
```

## Задача 4

Для данных из задачи 3 обучите случайный лес на 500 деревьев.

```python
n_estimators = 500
clf = RandomForestRegressor(n_estimators=n_estimators, random_state=42)
clf.fit(X_train, y_train)
```

Постройте график зависимости ошибки модели в зависимости от количества деревьев для обучающей и валидационной выборок.

*Замечание.* Не нужно заново обучать лес каждый раз. Используйте у *обученной* модели поле `estimators_`, это список из объектов класса `DecisionTreeRegressor`. Обратите внимание, что это приватное поле объекта, и менять его очень плохо, за такое баллы будут снижаться.

```python
mape_val = np.zeros(n_estimators)
mape_train = np.zeros(n_estimators)
tmp_pred_val = np.zeros(y_valid.shape)
tmp_pred_train = np.zeros(y_train.shape)
for i in range(500):
    tmp_pred_val += clf.estimators_[i].predict(X_valid)
    mape_val[i] = mean_absolute_percentage_error(y_valid, tmp_pred_val/(i+1))
    tmp_pred_train += clf.estimators_[i].predict(X_train)
    mape_train[i] = mean_absolute_percentage_error(y_train, tmp_pred_train/(i+1))
```

```python
plt.plot(range(1,501), mape_val, label = 'validation')
plt.plot(range(1,501), mape_train, label = 'train')
plt.xlabel('Number of trees')
plt.ylabel('MAPE error')
plt.legend();
```
