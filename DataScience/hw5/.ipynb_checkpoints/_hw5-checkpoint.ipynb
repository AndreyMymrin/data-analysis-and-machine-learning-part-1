**2.** Пусть есть выборка $X_1, ..., X_n$. Опишите процедуру бутстрепа построения доверительного интервала для величины $\theta = \mathsf{E} X_1$. Рассмотрите три типа  бутстрепных доверительных интервалов, выпишите формулу для каждого из них.

**Описание:**
1) выбираем число B

2) для каждого b: 1 <= b <= B генерируем случайные индексы {$i_{1}...i_{n}$} и получаем $X^{*}_{b} = (X_{i_{b1}}...X_{i_{bn}})$

3) Далее по каждой выборке посчитаем значение статистики T, получив выборку значений $T^*_b = T(X^*_b)$

4) Считаем бустрепную оценку дисперсии: $\widehat{V}_{boot} = \frac{1}{B}\sum_{b=1}^B (T^*_b)^2 - (\frac{1}{B}\sum_{b=1}^B T^*_b)^2$

- **Нормальный интервал**

Пусть $\widehat{\theta}$ — а.н.о. $\theta$ с ас. дисп. $\sigma^2(\theta)$.
$\widehat{v}_{boot}$ — бутстрепная оценка дисперсии.
Бутстрепный дов. интервал для параметра $\theta$ имеет вид:

$(\widehat{\theta} \pm z_{\frac{1+\alpha}{2}}\sqrt{\widehat{v}_{boot}})$

- **Центральный интервал**

Пусть $\theta$ = G(P) и $\widehat{\theta}$ = G($\widehat{P}_n$) — оценка методом подстановки.
$\theta^*_1,..., \theta^1_B$ — оценки по бутстрепным выборкам.
Бутстрепный доверительный интервал имеет вид

$C^* = (2\widehat{\theta} - \theta^*_{(\lceil B(1+\alpha)/2\rceil)}, 2\widehat{\theta} - \theta^*_{(\lfloor B(1-\alpha)/2\rfloor)})$

- **Квантильный интервал**

$\widehat{\theta}$ — некоторая оценка $\theta$.
$\theta^*_1,..., \theta^1_B$ — оценки по бутстрепным выборкам.
Бутстрепный доверительный интервал имеет вид:

$C^* = (\theta^*_{(\lfloor B(1-\alpha)/2\rfloor)}, \theta^*_{(\lceil B(1+\alpha)/2\rceil)})$

**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$. 

```python
def estimator(sample):
    '''
    Делает необходимые преобразования с выборкой (выполняет подсчет статистики).
    Используется в функции bootstrap для вычисления оценок.
    :param sample: массив с числами
    :return: оценка, посчитанная по выборке
    '''
    return sample.mean(axis = 1)

def bootstrap(sample, B=100_000):
    '''
    Реализует процедуру бутстрепа.
    :param sample: исходная выборка, по которой будут вычисляться бутстрепные оценки
    :param B: число бутстрепных оценок
    :return: возвращает массив длины B c бутстрепными оценками
    '''
    n = len(sample)  # длина выборки
    i_array = np.random.randint(0, n, (B, n))       
    X = np.array([[sample[i] for i in i_array[j]] for j in range(B)])
    return estimator(X)
```

**4.** Реализуйте три типа бутстрепных доверительных интервалы в виде функций, принимающих на вход выборку оценок, полученных с помощью бутстрепа. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки. 

```python
def boostrap_normal_confidence_interval(bootstrap_estimations,mean, alpha=0.95):
    return np.array([mean-sps.norm.ppf((1+alpha)/2)*np.sqrt(bootstrap_estimations.var()),\
                     mean+sps.norm.ppf((1+alpha)/2)*np.sqrt(bootstrap_estimations.var())])

    
def boostrap_central_confidence_interval(bootstrap_estimations,mean, alpha=0.95):
    sort_boot = np.array(sorted(bootstrap_estimations))
    B = len(bootstrap_estimations)
    
    return np.array([2*mean-sort_boot[math.ceil(B*(1+alpha)/2)],\
                     2*mean-sort_boot[math.floor(B*(1-alpha)/2)]])

    
def boostrap_quantile_confidence_interval(bootstrap_estimations, alpha=0.95):
    sort_boot = np.array(sorted(bootstrap_estimations))
    B = len(bootstrap_estimations)
    return np.array([sort_boot[math.floor(B*(1-alpha)/2)],\
                     sort_boot[math.ceil(B*(1+alpha)/2)]])
```

### Задача 6.


Сгенерируйте выборку $X = \left(X_1, \dots, X_{1000} \right)$ из стандартного распределения Коши. Для всех $n \leqslant 1000$ по первым $n$ элементам выборки $X_1, \dots, X_n$ вычислите значения следующих оценок:
- $\overline{X}$  —  выборочное среднее;
- $\overline{X}_{\alpha}$ —  усеченное среднее, где $\alpha$ &mdash; значение, на котором достигается минимум $\sigma^2_{\alpha}$;
- $\widehat{\mu}$ —  выборочная медиана;

а также, по каждой из этих оценок, одношаговую оценку. 

*Напоминание:* если $\widehat{\theta}_0$ — асимптотически нормальная оценка, то одношаговая оценка $\widehat{\theta}_1$ вычисляется как $\widehat{\theta}_1 = \widehat{\theta}_0 - \left(\ell''_X(\widehat{\theta}_0) \right)^{-1} \ell'_X(\widehat{\theta}_0)$, где $\ell_X(\theta)$ — логарифмическая функция правдоподобия. Заметим, что обычное выборочное среднее не является асимптотически нормальной оценкой, и оценка, вычисленная по формуле выше, формально не является одношаговой, однако ее все равно требуется посчитать.

На одном графике изобразите зависимость значений этих оценок от $n$. Для каждой оценки $\widehat{\theta}_0$ соответствующая оценка $\widehat{\theta}_1$ должна быть изображена на графике пунктиром тем же цветом, что и $\widehat{\theta}_0$. 

Сравните асимптотические дисперсии оценок. 

*Замечание:* если некоторые оценки имеют большой разброс, и разница между графиками зависимостей оценок с малыми значениями недостаточно заметна, стоит сделать два графика, на одном из которых будут изображены все оценки, а на втором &mdash; только достаточно хорошие.


$\alpha$ = 0.105 <- прошлое дз

$l_x(\theta) = -nln\pi - n -\sum(x_i-\theta)^2$

$l'_{x}(\theta) = 2\sum_{i = 1}^n(\theta-x_i)$

$l''_x(\theta) = 2n$

```python
sample_size = 1000
alpha = 0.105
sample_all = sps.cauchy.rvs(size = sample_size)
n = np.arange(1, sample_size,1)
```

```python
mean = sample_all.cumsum()[1:]/n
mean_trim = np.zeros(sample_size-1)
median = np.zeros(sample_size-1)
for i in range(0,sample_size-1):
    mean_trim[i] = sps.trim_mean(sample_all[0:i+1], alpha )
    median[i] = np.median(sample_all[0:i+1])
```

```python
median1 = np.zeros(sample_size-1)
mean1 = np.zeros(sample_size-1)
mean_trim1 = np.zeros(sample_size-1)
```

```python
for i in range(1,sample_size):
    sample = sample_all[:i]
    median1[i-1] = median[i-1] + ((sample-median[i-1])/(1+(sample-median[i-1])**2)).sum()/\
    ((1-(sample-median[i-1])**2)/((1+(sample-median[i-1])**2)**2)).sum() #формула с лекции
    mean1[i-1] = mean[i-1] - (mean[i-1] - sample).sum()/i
    mean_trim1[i-1] = mean_trim[i-1] - (mean_trim[i-1] - sample).sum()/i
```

```python
plt.figure(figsize=(15, 10))
plt.plot(n,mean_trim,'b', label = r'$\overline{X}_{\alpha}$', alpha = 0.5,linewidth=0.5)
plt.plot(n,mean,'r', label = r'$\overline{X}$', alpha = 0.5,linewidth=0.5)
plt.plot(n,median,'g', label = r'$\widehat{\mu}$', c = 'g',linewidth=0.5)
plt.plot(n,mean_trim1,'b--', label = r'$\overline{X}_{\alpha 1}$',linewidth=0.5)
plt.plot(n,mean1,'r--', label = r'$\overline{X_1}$',linewidth=0.5)
plt.plot(n,median1,'g--', label = r'$\widehat{\mu 1}$',linewidth=0.5)
plt.legend()
plt.ylim((-1,1));
plt.xlabel(r'n');
```

```python
plt.figure(figsize=(15, 7))
plt.plot(n,mean_trim,'b', label = r'$\overline{X}_{\alpha}$', alpha = 0.5,linewidth=0.5)
plt.plot(n,mean_trim1,'b--', label = r'$\overline{X}_{\alpha 1}$',linewidth=0.5)
plt.legend()
plt.xlabel(r'n');
plt.ylim((-0.1,0.1));
```

```python
plt.figure(figsize=(15, 7))
plt.plot(n,median,'g', label = r'$\widehat{\mu}$', c = 'g',linewidth=0.5)
plt.plot(n,median1,'g--', label = r'$\widehat{\mu 1}$',linewidth=0.5)
plt.legend()
plt.xlabel(r'n');
plt.ylim((-0.05,0.2));
```
